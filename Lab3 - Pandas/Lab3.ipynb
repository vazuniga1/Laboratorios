**Respuesta:**

````python
def clean_data(dataframe):
    """ 
    Limpia el dataframe de valores nulos, fuera de rango y duplicados.

    Args:
        dataframe: DataFrame a limpiar

    Returns:
        DataFrame limpio
    """
    # Se reemplazan valores vacios por NaN
    dataframe = dataframe.replace('', pd.NA)

    # Se eliminan registros con valores nulos
    dataframe = dataframe.dropna()

    # Se eliminan valores fuera del rango valido (0-100) para todas las columnas de notas
    dataframe = dataframe[(dataframe['math score'] >= 0) & (dataframe['math score'] <= 100)]
    dataframe = dataframe[(dataframe['reading score'] >= 0) & (dataframe['reading score'] <= 100)]
    dataframe = dataframe[(dataframe['writing score'] >= 0) & (dataframe['writing score'] <= 100)]

    # Se eliminan valores duplicados (un alumno dando el mismo examen dos veces)
    dataframe = dataframe.drop_duplicates(subset=['names'])

    return dataframe
````

Para una correcta limpieza de datos, se debe eliminar primero los valores fuera de rango ya que se permite limpiar los datos de manera m치s precisa. Por ejemplo, si se tienen dos filas duplicadas pero una de ellas contiene valores fuera de rango (o inv치lidos), eliminar los duplicados primero podr칤a implicar perder una fila con informaci칩n v치lida. Por lo que de cierta manera al eliminar primero los valores fuera de rango, se asegura conservar los datos correctos antes de manejar los duplicados.

### 3.1 Transformaci칩n de notas a "Nota Chilena" [0.3 Puntos]

En una de las reuniones con Don Caguayo este les comenta su intenci칩n de pasar todas sus notas a la "Escala Chilena". Para esto les facilita la siguiente ecuaci칩n:

$$Nota \, Chilena= \% \, de \, logro*0.06 + 1$$

Utilizando la ecuaci칩n entregada, cree una funci칩n que le permita pasar a la escala Chilena las notas del dataframe y entregue las notas redondeadas al segundo decimal.

**To-Do**

- [ ] Tranforme los valores a la escala chilena.

**Respuesta:**

````python
def nota_chilena(dataframe_in, columna):
    """
    Convierte las notas de una columna a escala chilena.
    
    Args:
        dataframe_in: DataFrame con las notas a convertir
        columna: Nombre de la columna con las notas a convertir
    
    Returns:
        DataFrame con las notas convertidas
    """
    # Se copia dataframe
    df = dataframe_in.copy()
    
    # Se convierten las notas a la escala chilena
    df[columna] = df[columna].apply(lambda x: (x*0.06) + 1)
    
    return df
````

### 3.2 Obtenci칩n de promedio de alumnos [0.3 Puntos]

**To-Do:**
- [ ] Crear funci칩n que nos permite obtener el promedio de cada uno de los alumnos y anexar esta informaci칩n en el `DataFrame` original.


**Respuesta:**

````python
def alumnos_promedio(dataframe_in):
    """
    Calcula el promedio de las notas de los alumnos y agrega una columna con el promedio.
    Args:
        dataframe_in: DataFrame
    Returns:
        DataFrame con una columna adicional llamada 'mean_test'
    """
    # Copiamos dataframe
    df = dataframe_in.copy()
    
    # Se calcula el promedio de las notas
    df['GPA'] = df[['math score', 'reading score', 'writing score']].mean(axis=1).round(2)
    
    return df
````

### 3.3 Reprobados y Destacados [0.9 Puntos, 0.3 cada DataFrame]

Contentos con su desempe침o, la Universidad les solicita que obtengan los alumnos quienes han reprobado el a침o y los que han tenido desempe침o de destacado. Para esto se considerar치 que un alumno ha reprobado el a침o cuando el promedio de notas es inferior a 4.0 en la Escala Chilena, y se considerar치 destacado si posee notas superiores a 5.8 en todas las asignaturas (o sea notas superiores a 5.8 en `math score`, `reading score` y `writing score`).

Cree una funci칩n que permita obtener tres Dataframes:
- El primero deber치 poseer solamente los alumnos que han reprobado en la prestigiosa universidad.
- El segundo debe entregar un DataFrame similar al original, pero marcado con una "R" el promedio de notas de los alumnos reprobados.
- El tercero ser치 un Data Frame con los alumnos destacados ordenados de mayor a menor.

**To-Do:**

- [ ] Obtener DataFrame con los alumnos que reprobaron el a침o.
- [ ] Obtener DataFrame similar al original, pero marcados con una "R" el atributo GPA de los alumnos reprobados.
- [ ] Obtener DataFrame con los alumnos destacados ordenados de mayor a menor.

**Respuesta:**

````python
def reprobados_destacados(dataframe_in):
    # Se obtiene dataframe con alumnos reprobados (promedio de notas menor a 4)
    df_reprobados = dataframe_in[(dataframe_in['GPA'] < 40)]

    # Se obtiene dataframe similar al original, pero marcados con R el atributo GPA de los reprobados
    df_rwr = dataframe_in.copy()
    df_rwr['GPA'] = df_rwr['GPA'].apply(lambda x: 'R' if x < 4.0 else x)

    # Se obtiene dataframe con alumnos destacados (nota superior a 58 en todas las asignaturas) y ordenados de mayor a menor
    df_dest = dataframe_in.copy()
    df_dest = df_dest[(df_dest['math score'] > 58) & (df_dest['reading score'] > 58) & (df_dest['writing score'] > 58)]
    df_dest = df_dest.sort_values(by='GPA', ascending=False)

    return df_reprobados, df_rwr, df_dest
````

**Respuesta:**

````python
# Calcular el promedio de las notas y guardarlo en una variable `GPA` (grade point average).
df_grades['GPA'] = df_grades[['math score', 'reading score', 'writing score']].mean(axis=1).round(2)

# Se realiza una simplificaci칩n a trav칠s de un mapeo de la columna 'parental level of education'.

# Se define el mapeo
mapeo = {'some high school': 'school',
    'some college': 'school',
    'high school': 'school',
    "bachelor's degree": 'college',
    "associate's degree": 'college',
    "master's degree": 'postgraduate'}

# Se agrega la columna respectiva y se hace el mapeo
df_grades['simple parental level of education'] = df_grades['parental level of education'].map(mapeo)

# Agregar seg칰n `race/ethnicity` y `simple parental level of education` para obtener el promedio de las notas.
df_grades_grouped = df_grades.groupby(['race/ethnicity', 'simple parental level of education'])[['math score', 'reading score', 'writing score']].mean().round(2)

# Agregar seg칰n 2 niveles: `race/ethnicity` y `simple parental level of education` para obtener conteo de alumnos por grupo
df_grades_grouped['count'] = df_grades.groupby(['race/ethnicity', 'simple parental level of education']).size()

# Obtener el porcentaje de alumnos con respecto al total. 
df_grades_grouped['percentage'] = (df_grades_grouped['count'] / df_grades_grouped['count'].sum() * 100).apply(lambda x: f'{x}%')
````
