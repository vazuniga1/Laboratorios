````python
# 3. Genere un ColumnTransformer en donde especifique las transformaciones que hay que realizar para cada columna
# (por ejemplo StandarScaler, MinMaxScaler, OneHotEncoder, etc...) para que puedan ser utilizadas correctamente por
# el modelo predictivo y gu√°rdelo una variable llamada col_transformer.

# Dividimos columnas por tipo de dato
categorical_data = ["Work_Rate", "Club_Position", "Preffered_Foot"]
numeric_standard =  numeric
numeric_minmax = ["Weak_foot", "Skill_Moves"]

# Definimos el ColumnTransformer
col_transformer = ColumnTransformer(
    transformers=[("categorical", OneHotEncoder(), categorical_data),
                  ("numerical_standard", StandardScaler(), numeric_standard ),
                  ("numerical_minmax", MinMaxScaler(), numeric_minmax)])
````

El StandardScaler se utiliza para variables que presentan una distribuci√≥n Gaussiana, un minmax para aquellas variables que tienen valores acotados y OneHotEncoder con las variables que son del tipo string ya que se pueden convertir a valores numericos.

Para lograr esto, primero, les pide que etiqueten de la siguiente manera los valores que aparecen en el atributo `Club_Position`, pidiendo que agrupen los valores en los siguientes grupos:

**Nota**:  Renac√≠n les recalca que **no deben utilizar los valores ```Sub``` y ```Res``` de esta columna**.


La elecci√≥n del clasificador se justificar en base a la siguiente [gu√≠a](https://scikit-learn.org/stable/tutorial/machine_learning_map/index.html) y se deben comentar los resultados obtenidos en la clasificaci√≥n.

**Tareas:** [1 punto por tarea]

1. En un nuevo dataframe, aplique las etiquetas descritas anteriormente en cada uno de los valores se√±alados en esta secci√≥n y gu√°rdelos en la variable `label`.

3. Entrene el nuevo pipeline y ejecute una evaluaci√≥n de este.  
4. Comente los resultados obtenidos.

**Respuesta:**

````python
# 1. Se definen las etiquetas en base a los distintos grupos
groups = {'ST': "ataque", 'CF': "ataque",
          'RW': "central_ataque" , 'CAM': "central_ataque", 'LW': "central_ataque",
          'RM': "central", 'CM': "central", 'LM':  "central",
          'RWB': "central_defensa", 'CDM': "central_defensa", 'LWB': "central_defensa",
          'RB': "defensa", 'CB': "defensa", 'LB': "defensa",
          'GK': "arquero"}

# Se hace la transformaci√≥n
label = df["Club_Position"].map(groups)
df["Club_Position"] = label

# 2. Se cuenta la cantidad de datos por etiqueta
label_count = label.value_counts()
print(label_count)
````

````python
# Se separa la data
df = df.dropna(subset=["Club_Position"])
X = df.drop(["Club_Position"], axis=1)
label = df["Club_Position"]

# Se crean conjuntos de entrenamiento
X_train, X_test, y_train, y_test = train_test_split(X, label, test_size=0.3, random_state=42, stratify=label)

# Se transforma la data
categorical_data = ["Work_Rate", "Preffered_Foot"]
numeric_standard =  [category for category in df.columns if df[category].dtype in ['int64', 'float64'] and category not in ["Weak_foot", "Skill_Moves"]]
numeric_minmax = ["Weak_foot", "Skill_Moves"]

col_transformer = ColumnTransformer(
    transformers=[("categorical", OneHotEncoder(), categorical_data),
                  ("numerical_standard", StandardScaler(), numeric_standard ),
                  ("numerical_minmax", MinMaxScaler(), numeric_minmax)])

pipeline_svm = Pipeline([
    ('transform', col_transformer),
    ('classifier', LinearSVC())
])

# Se entrena el pipeline
pipeline_svm.fit(X_train, y_train)

# Se evalua
pred_svm = pipeline_svm.predict(X_test)
print(classification_report(y_test, pred_svm))
````

De los resultados obtenidos anteriormente, es posible observar que el modelo predice muy bien para las posiciones de arquero y defensa, siendo seguido por ataque. Por otro lado, es posible observar que para las posiciones centrales no se obtienen los mejores resultados para ninguna de las m√©tricas, siendo f1-score la peor de ellas.

### 2.3 Calibraci√≥n Probabilistica [6 puntos]

<center>
<img src='https://media2.giphy.com/media/l2Je4Ku0Cx292KWv6/200w.gif?cid=6c09b952y0sihtq9tb6sz8j2023x3zxxp3qx1ocgonkpkblj&ep=v1_gifs_search&rid=200w.gif&ct=g' width=400 />

Para lograr modelos m√°s modulares, se recomienda realizar una calibraci√≥n del modelo entrenado anteriormente, con el objetivo de obtener salidas que reflejen mayor modularidad.

1. Se solicita que utilice un m√©todo de calibraci√≥n que asegure que las probabilidades generadas incrementen de manera mon√≥tona. Una m√©trica ampliamente utilizada para evaluar la precisi√≥n de la calibraci√≥n de un modelo es el Brier Score. Calcule el Brier Score para el modelo tanto antes como despu√©s de la calibraci√≥n. Esto le permitir√° realizar una comparaci√≥n cuantitativa y determinar si la calibraci√≥n ha mejorado el rendimiento del modelo. Para m√°s informaci√≥n sobre el Brier Score, puede consultar el siguiente enlace: [Scikit-Learn - Brier Score Loss](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.brier_score_loss.html). [3 puntos]

2. Tras la calibraci√≥n, examine y comente los resultados obtenidos. A su an√°lisis a√±ada una comparaci√≥n visual de las ideales versus las salidas del modelo original (sin calibrar) y del modelo calibrado. [3 puntos]

**Respuesta**

````python
from sklearn.metrics import brier_score_loss
from sklearn.calibration import CalibratedClassifierCV

# Predicci√≥n de probabilidad y brier score antes de la calibraci√≥n
y_test_prob_antes = model.predict_proba(X_test)
brier_score_antes = brier_score_loss(y_test, y_test_prob_antes[:, 1])

# Calibraci√≥n del modelo
calibrated_model = CalibratedClassifierCV(model, method='isotonic', cv='prefit')
calibrated_model.fit(X_train, y_train)

# Predicci√≥n de probabilidad y brier score despu√©s de la calibraci√≥n
y_test_prob_calibrated = calibrated_model.predict_proba(X_test)
brier_score_calibrated = brier_score_loss(y_test, y_test_prob_calibrated[:, 1])

# Comparaci√≥n antes y despu√©s
print("Brier Score antes de la calibraci√≥n:", brier_score_antes)
print("Brier Score despu√©s de la calibraci√≥n:", brier_score_calibrated)
````

````python
import matplotlib.pyplot as plt
from sklearn.calibration import calibration_curve
import plotly.graph_objs as go

# Se obtienen las curvas de calibraci√≥n para el modelo original y el modelo calibrado
prob_true_original, prob_pred_original = calibration_curve(y_test, y_test_prob_antes[:, 1], n_bins=10)
prob_true_calibrated, prob_pred_calibrated = calibration_curve(y_test, y_test_prob_calibrated[:, 1], n_bins=10)

fig = go.Figure()
fig.add_trace(go.Scatter(x=[0, 1], y=[0, 1], mode='lines', name='Ideal', line=dict(dash='dash')))
fig.add_trace(go.Scatter(x=prob_pred_original, y=prob_true_original, mode='markers+lines', name='Antes de calibrar', marker=dict(symbol='circle')))
fig.add_trace(go.Scatter(x=prob_pred_calibrated, y=prob_true_calibrated, mode='markers+lines', name='Despu√©s de calibrar', marker=dict(symbol='circle')))
fig.update_layout(title='Curvas de Calibraci√≥n',
                  xaxis_title='Probabilidad media prevista',
                  yaxis_title='Fracci√≥n de positivos')
fig.show()
````

Comentarios:
Es posible observar que el Brier Score antes de la calibraci√≥n es de 0.075, mientras que despu√©s de la calibraci√≥n aumenta a 0.085. Dado que un Brier Score m√°s alto indica un peor rendimiento en las predicciones, podemos concluir que la calibraci√≥n no mejor√≥ el modelo, ya que el valor empeor√≥ tras aplicarlo.

Lo anterior dicho puede verse reflejado tambi√©n en el gr√°fico, donde la curva del modelo sin calibrar est√° mas cerca de la curba ideal en comparaci√≥n con la curva posterior a la calibraci√≥n.
